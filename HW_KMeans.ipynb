```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
from sklearn.cluster import KMeans
```

```python
x_pca = a.x_pca
```

```python
from yellowbrick.cluster import KElbowVisualizer
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k = (2,50))
visualizer.fit(x_pca)
visualizer.poof()
```

```python
kmeans = KMeans(n_clusters=visualizer.elbow_value_)
```

```python
k_fit = kmeans.fit(x_pca)
```

```python
kümeler = k_fit.labels_
plt.scatter(x_pca.iloc[:,0], x_pca.iloc[:,1], c = kümeler, s = 50, cmap = 'viridis')
merkezler = k_fit.cluster_centers_
plt.scatter(merkezler[:,0], merkezler[:,1], c = 'black', s = 50, cmap = 'viridis')
```
